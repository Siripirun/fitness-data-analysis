# Fitness Data Analysis: Activity, Calories & Performance

## Objective
This project analyses Fitbit activity data to explore the relationship between physical activity, calorie expenditure, and performance metrics.

## Data Preparation

- Loaded Fitbit daily activity dataset  
- Reviewed dataset structure and variables  
- Prepared data for analysis using pandas  

## Steps and Calories

This section explores the relationship between daily step count and calorie expenditure.

```python
plt.figure(figsize=(10,6))
plt.scatter(df["TotalSteps"], df["Calories"])
plt.title("Steps vs Calories Burned")
plt.xlabel("Total Steps")
plt.ylabel("Calories Burned")
plt.tight_layout()
plt.show()
```

**Insight:** Higher step counts are generally associated with increased calorie expenditure, indicating a strong relationship between physical activity and energy use.

```python
plt.figure(figsize=(10,6))
df["VeryActiveMinutes"].hist(bins=30)
plt.title("Distribution of Very Active Minutes")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

**Insight:** Activity intensity varies significantly across individuals, suggesting differences in fitness behaviour and lifestyle.


```python
plt.figure(figsize=(10,6))
plt.scatter(df["SedentaryMinutes"], df["Calories"])
plt.title("Sedentary Minutes vs Calories")
plt.xlabel("Sedentary Minutes")
plt.ylabel("Calories")
plt.tight_layout()
plt.show()
```

**Insight:** Sedentary behaviour does not always directly correspond to lower calorie burn, suggesting multiple factors influence energy expenditure.

```python
plt.figure(figsize=(10,6))
plt.scatter(df["VeryActiveMinutes"], df["Calories"])
plt.title("Very Active Minutes vs Calories")
plt.xlabel("Very Active Minutes")
plt.ylabel("Calories")
plt.tight_layout()
plt.show()
```

**Insight:** Higher levels of intense activity are associated with increased calorie expenditure, highlighting the importance of exercise intensity.

## Key Insights

- Physical activity (steps and active minutes) is strongly linked to calorie expenditure  
- Individuals vary widely in activity intensity and fitness behaviour  
- Both movement and intensity contribute to overall performance outcomes  

## Conclusion

This analysis demonstrates how fitness data can be used to understand activity patterns and performance metrics. The results highlight the importance of both movement and exercise intensity in influencing energy expenditure.
